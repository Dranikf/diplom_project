# Мои попытки решить задачу классфикации с помошью нейронной сети

```python
import numpy as np


import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import matplotlib.pyplot as plt
import plotly.graph_objects as go

import os
from sys import path

exec_str = os.getcwd()[:os.getcwd().find('prog/')+len('prog/')]

if not exec_str in path:
    path.append(exec_str)
    print('exec added to path')


from common import *
import funs_params as fp
from data_loader import load_data
from DS_common import visualisations
```

# Загрузка и подготовка данных для классификации

Загрузка

```python
X, Y = load_data("2dim_data_for_nn.csv")
```

Подготовка тензоров

```python
X_tens = torch.tensor(X.astype('float32'))
Y_tens = torch.tensor(Y.reshape(len(Y),1).astype('float32'))
```

# Запись нейронной сети

```python
class SimpleNet(nn.Module):

    def __init__(self):
        super(SimpleNet, self).__init__()
        self.fc1 = nn.Linear(2, 2)
        self.fc2 = nn.Linear(2, 1)

        self.fc1.weight = \
        torch.nn.parameter.Parameter(torch.tensor([[-1.8193, 4.6818], [-5.7335, 2.2301]]))
        self.fc2.weight = \
        torch.nn.parameter.Parameter(torch.tensor([[-3.3749, -5.6865]]))
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

simple_net = SimpleNet()
```

# Оптимизатор

```python
optimizer = optim.Adam(simple_net.parameters(), lr=0.001)
```

# Функция потерь

```python
loss_fn = torch.nn.MSELoss()
```

# Пробуем немного поучить нашу сетку

Первая запись обучающего алгоритма

```python
print('layer 1')
print(simple_net.fc1.weight)
print(simple_net.fc1.bias)
print('layer 2')
print(simple_net.fc2.weight)
print(simple_net.fc2.bias)

for t in range(10000):
    predict = simple_net(torch.tensor(X.astype('float32')))
    loss = loss_fn(predict, Y_tens)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

print('layer 1')
print(simple_net.fc1.weight)
print(simple_net.fc1.bias)
print('layer 2')
print(simple_net.fc2.weight)
print(simple_net.fc2.bias)
```

```python
# сумматорная функция ReLU нейрона
# предполагается, что X - np.array()
sum_fun = lambda X: \
    torch.matmul(simple_net.fc1.weight, torch.tensor(X.astype('float32'))) +\
    simple_net.fc1.bias
```

# Визуализация заложенных мной закономерностей и того как нейронная сеть их отловила

Вычисляем значения сумматорных функций для различных наблюдений и получение массива, который для каждой записи покажет какие из ReLU нейронов пропускают сигнал.

```python
sum_f_values = \
    (simple_net.fc1.weight.detach().numpy().dot(X.transpose()) +\
     simple_net.fc1.bias.detach().numpy().reshape(2,1)).transpose()

ReLU_decisions = (sum_f_values > 0)
```

### Отрисовка

```python
fig = plt.figure(figsize = [15,10])


plt.subplot(121)

line1_x = [-fp.a/fp.a1, -(fp.a+fp.a2)/fp.a1]
line2_y = [-fp.b/fp.b2, -(fp.b+fp.b1)/fp.b2]


# точки
draw_my_scatter(
    X[:,0], X[:,1], Y
)
# дискриминирующие линии
plt.plot(line1_x, [0, 1], 'black', linestyle = '--')
plt.plot([0,1], line2_y, 'black', linestyle = '-.')
# площади
plt.fill_between([0] + line1_x, [0, 0, 1], [1]*3,  hatch = '/', alpha = 0)
plt.fill_between([0,1], line2_y, [1,1],  hatch = '\\', alpha = 0)

plt.legend([
        'Класс1', 'Класс2', 
        '$c_1x_1+c_2x_2+c=0$',
        '$d_1x_1+d_2x_2+d=0$'
    ],
    fontsize = 15, loc = 4, framealpha = 1
)


plt.title('Заложенное разбиение', fontsize = 17)
plt.margins(x = 0, y = 0)

plt.subplot(122)

def groups_plot(condition, **kawargs):
    plt.scatter(X[:,0][condition], X[:,1][condition], **kawargs)   

# разбиения маркерами
groups_plot(ReLU_decisions[:,0] & ReLU_decisions[:,1], marker = 'X', s = 100)
groups_plot(ReLU_decisions[:,0] & np.invert(ReLU_decisions[:,1]), marker = '>', s = 100)
groups_plot(np.invert(ReLU_decisions[:,0]) & ReLU_decisions[:,1], marker = '<', s = 100)
groups_plot(np.invert(ReLU_decisions[:,0] | ReLU_decisions[:,1]), marker = '+', s = 100)

# ошибки разбиений
cond1 = \
    (fp.a1*X[:,0] + fp.a2*X[:,1] + fp.a > 0) &\
    ReLU_decisions[:,1]
cond2 = \
    (fp.b1*X[:,0] + fp.b2*X[:,1] + fp.b > 0) &\
    ReLU_decisions[:,0]
plt.scatter(X[cond1,0], X[cond1,1], facecolors='none', edgecolors='black', s = 140)
plt.scatter(X[cond2,0], X[cond2,1], facecolors='none', edgecolors='black', s = 140)


plt.title("Предсказонное разбиение", fontsize = 17)
plt.xlabel("$x_1$", fontsize = 15)

plt.legend([
        "$z_1^1>0;z_2^1>0$",
        "$z_1^1>0;z_2^1<0$",
        "$z_1^1<0;z_2^1>0$",
        "$z_1^1<0;z_2^1<0$",
        "Ошибки"
    ], 
    fontsize = 15,  
    framealpha = 1,
    loc = 4
)

fig.tight_layout()
plt.savefig('pictures/non_linear_for_nn.png')
```

# Нанесем трехмерный график предсказания вероятностей

Подкотовка данных для рисунка

```python
mesh_ax_points = 100

x1 = np.linspace(-0.5, 1.5, mesh_ax_points)
x2 = np.linspace(-0.5, 1.5, mesh_ax_points)

x1, x2 = np.meshgrid(x1,x2)
i_ind, j_ind, k_ind = visualisations.get_meshtraces(x1)
pp = simple_net(
    torch.tensor(
        np.concatenate(
            [
                x1.reshape(mesh_ax_points**2,1),
                x2.reshape(mesh_ax_points**2,1)
            ], 
            axis = 1
        ).astype('float32')
    )
).detach().numpy()
```

Отрисовка графика

```python

data = []

# подготовка элементов графика
data.append(
    go.Scatter3d(
        x = X[np.invert(Y), 0].flatten(),
        y = X[np.invert(Y), 1].flatten(),
        z = probs_hat[np.invert(Y)].flatten(),
        marker = {'size':5, 'color':'#0000ff'},
        name = 'класс1',
        mode = 'markers',
    )
)

data.append(
    go.Scatter3d(
        x = X[Y, 0].flatten(),
        y = X[Y, 1].flatten(),
        z = probs_hat[Y].flatten(),
        marker = {'size':5, 'color':'red'},
        name = 'класс2',
        mode = 'markers'
    )
)


data.append(
    go.Mesh3d(    
        x=x1.ravel(),
        y=x2.ravel(),
        z=pp.ravel(),
        opacity=0.5, color='gray',
        colorbar_thickness=25, 
        flatshading=True,
        showlegend = True,
        name = 'сигмоида',
        i = i_ind, j = j_ind, k = k_ind
    )
)

print(pp.ravel().max())

# настройка отображения
fig = go.Figure(data)

fig.update_layout(
    height=750, width=1500
)

fig.layout = go.Layout(
    scene = dict(
        xaxis = dict(title='x',titlefont = {'size':20}),
        yaxis = dict(title='x',titlefont = {'size':20}),
        zaxis = dict(title='p',titlefont = {'size':20}),
    ),
    legend=dict(
        font=dict(
            family="Courier",
            size=20,
            color="black",
        )
    ),
    width=1500,
    height=1500
)


fig.show()
```
