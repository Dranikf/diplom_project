# Попытки получить информацию о клиентах, которым старые модели не выдали кредит

```python
import pandas as pd
import os
import os.path as os_path
import numpy as np

from sys import path

# current project file name
cpfn = "diplom_project"

curr_dir = os.getcwd()
curr_dir = os_path.join(curr_dir[:curr_dir.find(cpfn)], cpfn)
data_storage = os_path.join(curr_dir, "data_storage")
input_file_dir = os_path.join(
    os_path.join(data_storage, "scoring_result"), "2022")
proc2_model_fp = os_path.join(data_storage, "proc_2")
output_dir = os_path.join(data_storage, "proc_old_model")

if curr_dir not in path:
    path.append(curr_dir)

from data_description import get_data_frame_settings
```

# Извлечение информации и первичная обработка

```python
# загрузка
data = pd.read_excel(input_file_dir + ".xls", skiprows = 17)
data.drop(data.iloc[:, [0,3,4]].columns, axis = 1, inplace = True)
data.drop('Дата регистрации договора', axis = 1, inplace = True)

# переименования столбцов
data.rename(columns = {"Unnamed: 1" : "Дата регистрации договора", 
                       "Unnamed: 2": "Код подразделения"}, inplace= True)
data.columns = data.columns.str.replace(".._", "", regex = True)
```

# Детальная подгонка данных к используемым в модели

Нужно повторить преобразования

```python
data['Срок кредита в днях'] = '-'
data['Y'] = (data['Дефолт в любом периоде'] > 0.67).astype('int32')
```

Подгрузим преобразованные данные для модели, и вытянем от туда самые популярные занчения для тех переменных которые для старой модели не заполнялись.

```python
proc2_data = pd.read_csv(proc2_model_fp + ".csv", index_col = 0)

for col in proc2_data.columns:
    if col == "Социальная группа":
        a = 5
    
    if (data[col] == '-').all():
        if proc2_data[col].dtype == 'O':
            data[col] = proc2_data[col].value_counts().index[0]
        else:
            data[col] = proc2_data[col].median()
            
data = data.iloc[:, data.columns.isin(proc2_data.columns)]
data.replace('-', 'нет данных', inplace = True)
```

Доподгонка вручную

```python
data['Гражданин РБ'] = "РБ"

# удаление скробок и замено пропусков у в реальности численных переменных
def replace_brakets(column):
    return  column.str.replace(" \[(.+)\]", "", regex = True).\
            astype('int').replace({-1: np.NaN})

rep_cols = ['Количество действующих кредитных договоров',
           'Максимальное количество дней просрочки',
           'Количество запросов в КБ за последние 30 дней']

data[rep_cols] = data[rep_cols].apply(replace_brakets, axis = 1)
```

```python
data.loc[data['Количество запросов в КБ за последние 30 дней'] > 2,
              'Количество запросов в КБ за последние 30 дней_'] = '>2.0'
data.loc[data['Количество запросов в КБ за последние 30 дней'] <= 2,
              'Количество запросов в КБ за последние 30 дней_'] = '<=2.0'

data['Количество запросов в КБ за последние 30 дней'] = \
data['Количество запросов в КБ за последние 30 дней_']

data.drop('Количество запросов в КБ за последние 30 дней_', axis = 1, inplace = True)
data['Количество запросов в КБ за последние 30 дней'].fillna('нет данных', inplace = True)
```

```python
data["Наличие кредитной истории"].replace({
    '1 [есть]' : 'есть', '0 [нет]': 'нет'
}, inplace = True)
```

```python
data.loc[data['Код подразделения'].isin(data['Код подразделения'].unique()[np.invert(np.isin(
    data['Код подразделения'].unique(),
    proc2_data['Код подразделения'].unique()
))]), 'Код подразделения'] = proc2_data['Код подразделения'].value_counts().index[0]
```

```python
data['Воинская служба'].replace(
    {'не военнообязанный' : 'невоеннообязанный'}, 
    inplace = True)
```

```python
data.loc[data['Недвижимость'] != 'нет', 'Недвижимость'] = 'есть'
```

```python
data.drop(
    data.loc[data['Работа занимаемая должность'] == 'сотрудник организации'].index,
    inplace = True)

data['Семейное положение'].replace(
     {'Женат/замужем':'женат/замужем',
      'в браке/повторный брак': 'повторный брак'}, inplace = True)
```

## Сверка данных

```python
get_data_frame_settings(data)[['Область допустимых значений']].join(
    get_data_frame_settings(proc2_data)[['Область допустимых значений']], lsuffix = "_old"
)
```

# Проверка, чтобы в категориях новых данных не содержалось ничего нового (чего модель не знает)

```python
def check(col_name):
    return np.all(np.isin(data[col_name].unique(), proc2_data[col_name].unique()))


res = {}
for col in data.loc[:, data.dtypes == 'O']:
    res[col] = check(col)
    
res
```

# Сохранение такого обновленного набора данных

```python
data[proc2_data.columns].to_csv(output_dir + '.csv')
```
