# Описание идеи ROC анализа

```python
from scipy.stats import norm
import numpy as np

from plotly.subplots import make_subplots
import plotly.graph_objects as go

from gifmaker.gifmaker import creategif

import os
import os.path as os_path
from os import listdir

cpfn = "diplom_project"

curr_dir = os.getcwd()
curr_dir = os_path.join(curr_dir[:curr_dir.find(cpfn)], cpfn)

# путь к этой книжке и результату выполнения
prog_fold = os_path.join(curr_dir, "prog")
defence_fold = os_path.join(prog_fold, "defence_works")
ROC_result_fold = os_path.join(defence_fold, "ROC_idea")
distrs_fold = os_path.join(ROC_result_fold, "diff_distrs_plots")
distrs_dir = os_path.join(distrs_fold, "")
```

# Базовые функции и настройки

## Функции

```python
basic_x = np.linspace(0, 1, 1000)

def get_pdf_frame(
    m, sigma, 
    x = basic_x
):
    return norm(m, sigma).pdf(x)

def get_cdf_frame(
    m, sigma,
    x = basic_x
):
    return norm(m, sigma).cdf(x)
```

## Настройки

```python
# цвет линий для различных классов
classes_colors = ["#00FF00", "#FF0000"]
```

# Подготовим автоматический построитель

```python
def build_ROC_fig(m1, m2, sigma):
    '''Построить распределение плотности и вероятности для двух классов'''
    # inputs:
    # m1 - математическое ожидание первого класса
    # m2 - математическое ожидание второго класса
    # sigma - СКО обоих распределений
    ROC_fig = make_subplots(rows = 2, cols = 1)
    
    # Плотности распределения вероятностей
    # ====================================================
    ROC_fig.add_trace(
        go.Scatter(
            x = basic_x,
            y = get_pdf_frame(m1, sigma),
            name = 'класс 0',
            line_color = classes_colors[0]
        ),
        row = 1, col = 1
    )
    ROC_fig.add_trace(
        go.Scatter(
            x = basic_x,
            y = get_pdf_frame(m2, sigma),
            name = 'класс 1',
            line_color = classes_colors[1]
        ),
        row = 1, col = 1
    )
    
    # Функции распределения вероятностей
    # ====================================================
    # кумулятивные функции распределения 
    # различных классов нам еще понадобятся
    classes_cdfs = [get_cdf_frame(m1, sigma), get_cdf_frame(m2, sigma)]
    
    ROC_fig.add_trace(
        go.Scatter(
            x = basic_x,
            y = classes_cdfs[0],
            line_color = classes_colors[0],
            showlegend = False
        ),
        row = 2, col = 1
    )

    ROC_fig.add_trace(
        go.Scatter(
            x = basic_x,
            y = classes_cdfs[1],
            line_color = classes_colors[1],
            showlegend = False
        ),
        row = 2, col = 1
    )
    
    # Добавление на график информации о KS статистике
    # ====================================================
    
    # надо узнать в какой дочке
    # дейсвует KS статистика
    KS_ind = np.argmax(classes_cdfs[0] - classes_cdfs[1])
    
    arrows_settings = dict(
        row = 2, col = 1,
        xref = "x2", yref = "y2",
        axref = "x2", ayref = "y2",
        arrowhead = 2,
        arrowwidth = 1.5
    )
    
    
    ROC_fig.add_annotation(
        x = basic_x[KS_ind], ax = basic_x[KS_ind],
        y = classes_cdfs[0][KS_ind], ay = classes_cdfs[1][KS_ind],
        **arrows_settings
    )
    ROC_fig.add_annotation(
        x = basic_x[KS_ind], ax = basic_x[KS_ind],
        y = classes_cdfs[1][KS_ind], ay = classes_cdfs[0][KS_ind],
        **arrows_settings
    )
    
    ROC_fig.add_annotation(
        x = basic_x[KS_ind] + 0.015, 
        y = (classes_cdfs[1][KS_ind] + classes_cdfs[0][KS_ind])/2,
        text = "KS",
        row = 2, col = 1,
        showarrow = False
    )
    
    # ====================================================
    
    
    # настройки осей
    ROC_fig.update_layout(
        width = 800,
        height = 600,
        yaxis_title = "$\hat{f}(p)$",
        yaxis2_title = "$\hat{F}(p)$",
        xaxis2_title = 'P'
    )
    
    return ROC_fig

build_ROC_fig(0.3, 0.6, 0.1)
```

# Подготовка гифки описывающей этот процесс

Постепенно отдаляем распределения двух классов, сохраняем для каждого варианта картинку а в конце концов, лепим из них гифку

```python
pic_count = 40

for i in range(pic_count):
    padding = (i + 1)*0.2/pic_count
    file_name = distrs_dir + str(i) + '.png'
    build_ROC_fig(0.5 - padding, 0.5 + padding, 0.1).write_image(file_name)
    
    
creategif(
    [os_path.join(distrs_fold, filename) for filename in listdir(distrs_fold)],
    ROC_result_fold + "/distrs.gif",0.1
)
```
